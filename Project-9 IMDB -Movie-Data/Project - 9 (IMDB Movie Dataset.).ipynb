# 22. Find Unique Values From Genre 


```python
data.columns
```

```python
data['Genre']
```

```python
# step -1 split items by commas

list1= []
for value in data['Genre']:
    list1.append(value.split(','))
```

```python
list1
```

```python
# step -2 converting 2d list to 1d list

one_d=[]
for item in list1:
    for item1 in item:
        one_d.append(item1)
```

```python
one_d
```

```python
# step -3 we will find unique values 

uni_list=[]
for item in one_d:
    if item not in uni_list:
        uni_list.append(item)
```

```python
uni_list  # this is the answer
```

```python
len(uni_list)
```
